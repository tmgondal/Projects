In the following cells, I have executed all the analysis


```python
df.isnull() #this will identify the Column with Null Values
```

In above cell it can be observed that "Alley, Fence, Pool QC" and other columns are showing True, which mean they have NaN Values. Lets get the name of all the columns with NaN Values

```python
null_list = df.columns[df.isnull().any()].tolist() #this will check all the columns for null valus and generate a list for them
null_list #shows the list of columns with null values
```

Now I have the list of all the columns with null values. But these are only the names of columns. I am not shure how many entries are there with null values. To obtain the number of null values, i need to sum all of those. so lets calculate the null total enteries.

```python
null_total = df[null_list].isnull().sum() # Sum of nan values of each column
null_total #prints the null values
```

Now, I have all the numbers, it can be estimated that Pool QC has the highest number of NaN values while few columns have only 1 nan entry. To better observe the trend of these values, lets plot a bart chart for these values.

```python
null_total.plot(kind='barh', title='Bar Chart for Null Values of Each Columns',figsize=(18,6),grid=True)
plt.show()
```

from above chart we visualize the total number of Null Values in each columns. Misc Feature, Fence, Poo QC, and Alley has the highest Null Values.
Now, lets further dive into by calculating the percentage of each missing value in each columns. After this we will remove these values.

```python
# Find how many missing values and percentage missing in each column
print('Missing Values')
for col in df.columns:
    missing = len(df) - df[col].count()
    if (missing > 0 or missing =='NaN'):
         print(col, ':', missing, 'missing values is', str(round(float(missing/float(len(df)))*100, 3)), '% of total')
```

Above cell is not necessarily required to complete the data cleaning but it provides the information regarding how many percent data is null. Now lets apply few methods to fill these null values.

**Method 1:**
Apply Interpolation

In a dataframe or series, the dataframe.interpolate() method of Pandas is mostly used to fill NA values. However, this feature can be used extremely well to fill in the blanks. Instead of hard-coding the value, it fills in the missing numbers using a variety of interpolation techniques.

```python
df=df.interpolate() #applying method
```

```python
null_total = df[null_list].isnull().sum() # Sum of nan values of each column
null_total #prints the null values
```

By applying interpolate method, it can be observed that there are still many entries with Null Value. So the method one has not effectively worked with this case. so now apply method 2.

**Method 2:**
Fillna()

Fillna() controls and enables the user to override NaN values with custom values.

```python
# Build a list of colums where NaN should be a string 'NA'
col_na = ['Mas Vnr Type','Bsmt Qual','Bsmt Cond','Bsmt Exposure','BsmtFin Type 1','BsmtFin Type 2','Electrical','Fireplace Qu',
          'Garage Type','Garage Finish','Garage Qual','Garage Cond','Fence','Pool QC','Alley','Misc Feature']
```

```python
df[col_na] = df[col_na].fillna('0') #replace all of those columns with zero
```

```python
# Create list of numeric columns
numeric_columns = df.select_dtypes(include=[np.number]).columns.tolist()
print(numeric_columns)
```

```python
# Draw scatter plots of numerical columns
import seaborn as sns #seaborn is also used for the visualization purposes. I have used it to show indepth analysis.
def draw_scatters(df, variables, n_rows, n_cols): # reference to this funciton is given in bibiliography
    fig=plt.figure(figsize=(15,60))
    for i, var_name in enumerate(variables):
        ax=fig.add_subplot(n_rows,n_cols,i+1)
        sns.regplot(x=var_name,y='SalePrice',data=df,fit_reg=False, scatter_kws={'alpha':0.2}) #selecting out dataset as input with SalePrice
        ax.set_title(var_name +" vs. Sale Price") #comparison with numeric columns
    fig.tight_layout()
    plt.show()
draw_scatters(df, numeric_columns[:-1], int(len(numeric_columns[:-1])/2)+1, 2) 
#will show results for ['Order', 'PID', 'Lot Frontage', 'Lot Area', 'Overall Qual', 'Overall Cond', 'Year Built', 
#'Year Remod/Add', 'Mas Vnr Area', 'BsmtFin SF 1', 'BsmtFin SF 2', 'Bsmt Unf SF', 'Total Bsmt SF', '1st Flr SF',
#'2nd Flr SF', 'Low Qual Fin SF', 'Gr Liv Area', 'Bsmt Full Bath', 'Bsmt Half Bath', 'Full Bath', 'Half Bath', 
#'Bedroom AbvGr', 'Kitchen AbvGr', 'TotRms AbvGrd', 'Fireplaces', 'Garage Yr Blt', 'Garage Cars', 'Garage Area',
# 'Wood Deck SF', 'Open Porch SF', 'Enclosed Porch', '3Ssn Porch', 'Screen Porch', 'Pool Area', 'Misc Val', 'Mo Sold', 'Yr Sold', 'SalePrice']
```

```python
#This will provide the mean and median values along with the histogram for Saleprice

df['SalePrice'].hist(bins=50, alpha=0.8)
plt.axvline(df['SalePrice'].mean(), color='r', linestyle='dashed', linewidth=2, label='Mean')
plt.axvline(df['SalePrice'].median(), color='orange', linestyle='dashed', linewidth=2, label='Median')
plt.xlabel('Sale Price')
plt.ylabel('Count')
plt.legend()
plt.show()
```

```python
#This will provide the mean and median values along with the histogram for Lot Area

df['Lot Area'].hist(bins=50, alpha=0.8)
plt.axvline(df['Lot Area'].mean(), color='r', linestyle='dashed', linewidth=2, label='Mean')
plt.axvline(df['Lot Area'].median(), color='orange', linestyle='dashed', linewidth=2, label='Median')
plt.xlabel('Lot Area')
plt.ylabel('Count')
plt.legend()
plt.show()
```
